```python
from tkinter import Tk, Label, Checkbutton, StringVar, IntVar, END, Button, Text, Scrollbar, RIGHT, Y, Entry

# Função 1 - Login Metamask
def login_metamask():
    caixa_texto.insert(END, "Login Metamask clicado!\n")
    caixa_texto.see(END)  # Rola automaticamente para a última mensagem

    # Desmarca as outras opções do Metamask
    opcao2_var.set(0)
    opcao3_var.set(0)
    opcao4_var.set(0)
    opcao5_var.set(0)

# Função 2 - Abrir Discord
def abrir_discord():
    caixa_texto.insert(END, "Abrir Discord clicado!\n")
    caixa_texto.see(END)  # Rola automaticamente para a última mensagem

    # Desmarca as outras opções do Discord
    Disc2_var.set(0)
    Disc3_var.set(0)
    Disc4_var.set(0)
    Disc5_var.set(0)

# Função para lidar com a seleção das caixas de seleção do Metamask
def handle_metamask_checkbox_selection():
    opcoes_selecionadas = []

    if opcao1_var.get() == 1:
        opcoes_selecionadas.append("10 Contas")

    if opcao2_var.get() == 1:
        opcoes_selecionadas.append("20 Contas")
        # Desmarca as outras opções do Metamask
        opcao1_var.set(0)
        opcao3_var.set(0)
        opcao4_var.set(0)
        opcao5_var.set(0)

    if opcao3_var.get() == 1:
        opcoes_selecionadas.append("30 Contas")
        # Desmarca as outras opções do Metamask
        opcao1_var.set(0)
        opcao2_var.set(0)
        opcao4_var.set(0)
        opcao5_var.set(0)

    if opcao4_var.get() == 1:
        opcoes_selecionadas.append("40 Contas")
        # Desmarca as outras opções do Metamask
        opcao1_var.set(0)
        opcao2_var.set(0)
        opcao3_var.set(0)
        opcao5_var.set(0)

    if opcao5_var.get() == 1:
        opcoes_selecionadas.append("50 Contas")
        # Desmarca as outras opções do Metamask
        opcao1_var.set(0)
        opcao2_var.set(0)
        opcao3_var.set(0)
        opcao4_var.set(0)

    if opcoes_selecionadas:
        mensagem = "Selecionado(s) do Metamask: " + ", ".join(opcoes_selecionadas)
    else:
        mensagem = "Nenhuma opção do Metamask selecionada"

    caixa_texto.insert(END, mensagem + "\n")
    caixa_texto.see(END)  # Rola automaticamente para a última mensagem

# Função para lidar com a seleção das caixas de seleção do Discord
def handle_discord_checkbox_selection():
    opcoes_selecionadas = []

    if Disc1_var.get() == 1:
        opcoes_selecionadas.append("10 Discords")

    if Disc2_var.get() == 1:
        opcoes_selecionadas.append("20 Discords")
        # Desmarca as outras opções do Discord
        Disc1_var.set(0)
        Disc3_var.set(0)
        Disc4_var.set(0)
        Disc5_var.set(0)

    if Disc3_var.get() == 1:
        opcoes_selecionadas.append("30 Discords")
        # Desmarca as outras opções do Discord
        Disc1_var.set(0)
        Disc2_var.set(0)
        Disc4_var.set(0)
        Disc5_var.set(0)

    if Disc4_var.get() == 1:
        opcoes_selecionadas.append("40 Discords")
        # Desmarca as outras opções do Discord
        Disc1_var.set(0)
        Disc2_var.set(0)
        Disc3_var.set(0)
        Disc5_var.set(0)

    if Disc5_var.get() == 1:
        opcoes_selecionadas.append("50 Discords")
        # Desmarca as outras opções do Discord
        Disc1_var.set(0)
        Disc2_var.set(0)
        Disc3_var.set(0)
        Disc4_var.set(0)

    if opcoes_selecionadas:
        mensagem = "Selecionado(s) do Discord: " + ", ".join(opcoes_selecionadas)
    else:
        mensagem = "Nenhuma opção do Discord selecionada"

    caixa_texto.insert(END, mensagem + "\n")
    caixa_texto.see(END)  # Rola automaticamente para a última mensagem

# Criação da janela principal
window = Tk()
window.title("Meganinos")

# Definindo o tamanho da janela
window.geometry("900x600")  # Tamanho: 900x600 pixels

# Impedindo que a janela seja redimensionada pelo usuário
window.resizable(False, False)

# Rótulo
label = Label(window, text="Olá, mundo!")
label.pack()

# Função 1 - Botão Login Metamask
botao_login_metamask = Button(window, text="Login Metamask", command=login_metamask)
botao_login_metamask.place(x=10, y=30)  # Definindo as coordenadas do botão

# Função 2 - Botão Abrir Discord
botao_abrir_discord = Button(window, text="Abrir Discord", command=abrir_discord)
botao_abrir_discord.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=30)

# Espaço reservado para a Função 2

# Menu de Opções com Caixas de Seleção para o Metamask
opcao1_var = IntVar()
opcao2_var = IntVar()
opcao3_var = IntVar()
opcao4_var = IntVar()
opcao5_var = IntVar()

opcao1_checkbox = Checkbutton(window, text="10 Contas", variable=opcao1_var, command=handle_metamask_checkbox_selection)
opcao1_checkbox.place(x=10, y=70)

opcao2_checkbox = Checkbutton(window, text="20 Contas", variable=opcao2_var, command=handle_metamask_checkbox_selection)
opcao2_checkbox.place(x=10, y=100)

opcao3_checkbox = Checkbutton(window, text="30 Contas", variable=opcao3_var, command=handle_metamask_checkbox_selection)
opcao3_checkbox.place(x=10, y=130)

opcao4_checkbox = Checkbutton(window, text="40 Contas", variable=opcao4_var, command=handle_metamask_checkbox_selection)
opcao4_checkbox.place(x=10, y=160)

opcao5_checkbox = Checkbutton(window, text="50 Contas", variable=opcao5_var, command=handle_metamask_checkbox_selection)
opcao5_checkbox.place(x=10, y=190)

# Menu de Opções com Caixas de Seleção para o Discord
Disc1_var = IntVar()
Disc2_var = IntVar()
Disc3_var = IntVar()
Disc4_var = IntVar()
Disc5_var = IntVar()

Disc1_checkbox = Checkbutton(window, text="10 Discords", variable=Disc1_var, command=handle_discord_checkbox_selection)
Disc1_checkbox.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=70)

Disc2_checkbox = Checkbutton(window, text="20 Discords", variable=Disc2_var, command=handle_discord_checkbox_selection)
Disc2_checkbox.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=100)

Disc3_checkbox = Checkbutton(window, text="30 Discords", variable=Disc3_var, command=handle_discord_checkbox_selection)
Disc3_checkbox.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=130)

Disc4_checkbox = Checkbutton(window, text="40 Discords", variable=Disc4_var, command=handle_discord_checkbox_selection)
Disc4_checkbox.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=160)

Disc5_checkbox = Checkbutton(window, text="50 Discords", variable=Disc5_var, command=handle_discord_checkbox_selection)
Disc5_checkbox.place(x=botao_login_metamask.winfo_x() + botao_login_metamask.winfo_width() + 130, y=190)

# Caixa de texto
caixa_texto = Text(window, width=40, height=35)
caixa_texto.pack(side="right")

# Caixa de link
link_entry = Entry(window, width=60)
link_entry.place(x=10, y=250)

# Barra de rolagem
scrollbar = Scrollbar(window)
scrollbar.pack(side=RIGHT, fill=Y)

# Vinculando a barra de rolagem com a caixa de texto
caixa_texto.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=caixa_texto.yview)

# Função 3
def funcao3():
    pass

# Espaço reservado para a Função 3

# Função 4
def funcao4():
    pass

# Espaço reservado para a Função 4

# Função 5
def funcao5():
    pass

# Espaço reservado para a Função 5

# Execução da interface
window.mainloop()
```
